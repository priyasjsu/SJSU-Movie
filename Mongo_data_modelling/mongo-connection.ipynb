### (MA1)This query allows manager to fetch customers whose bill is due on the present day

```python
collection_name=dbname['user']
agg_result= collection_name.aggregate([
    {
        '$match': {"billing_info.due_date":"2022-05-07T00:00:00"}
        },
        {
        '$project': {
            'name': 1, 
            'email': 1
        }
    }
])

for i in agg_result:
    print(i)
```

### (MA2)This aggregation allows manager to fetch details of monthly subscription numbers

```python
collection_name=dbname['user']
agg_result= collection_name.aggregate([{
    "$project": { "_id":0,
        'created_at': 1
    }
}, {
    "$match": {
        'created_at': {
            "$gte": '01-02-20',
            "$lte": '28-02-20'
        }
    }
},
{
    "$count": 'Number of new Customers in February'
}])

for i in agg_result:
    print(i)
```

### (MA4)This aggregation allows manager to fetch total number of problems faced by users in a month (e.g. July, 2021)

### MA5) This aggregation allows manager to fetch list of which employees resolved which customer service issue.

### CU1) This aggregation allows User to find out watched movies of specific profile using user email and profile name.

### MA10) Top 5 genres watched by users

```python
collection_name=dbname['user']
agg_result= collection_name.aggregate([{
 "$lookup": {
 "from": 'movies',
 "localField": 'profiles.history.watched_movie_no',
 "foreignField": '_id',
 "as": 'Movie'
}}, {"$unwind": {
 "path": '$Movie'
}}, {"$unwind": {
 "path": '$Movie.genres'
}}, {"$group": {
 "_id": '$Movie.genres',
 "count": {
  "$count": {}
 }
}}, {"$sort": {
 "count": -1
}}, {"$limit": 5}])


for i in agg_result:
    print(i)
```
